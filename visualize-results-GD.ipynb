```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import pickle


# define constants and other global variables
label_size = 28
tick_size = 20
title_size = 40
subtitle_size = 28
line_width = 3
legend_size = 20
```

```python
imp_df = pd.read_csv("results/CIFAR10_ResNet20/results/IMP/IMP_cifar10_resnet20_prune_0_2.csv")
```

```python
RESULTS_FOLDER = "results/CIFAR10_ResNet20/results/finetune_results"
sp_list = [1,2,3,4,5,6,8,10,13,16,20,26,32,40,51,63,79] #[1]#
vanilla_acc = []
mask_shuffle_acc = []
weight_shuffle_acc = []
#idx = [-10:-1]

weight_reinit_df = pd.read_csv("{}/finetune_reinit_weight_1.csv".format(RESULTS_FOLDER))
for sp in sp_list:

    vanilla_acc_df = pd.read_csv("{}/finetune_{}.csv".format(RESULTS_FOLDER, sp))
    mask_shuffle_df = pd.read_csv("{}/finetune_shuffle_mask_{}.csv".format(RESULTS_FOLDER, sp))
    weight_shuffle_df = pd.read_csv("{}/finetune_shuffle_weight_{}.csv".format(RESULTS_FOLDER, sp))
    
    vanilla_acc.append(np.mean(vanilla_acc_df['test_acc'][-10:-1]))
    mask_shuffle_acc.append(np.mean(mask_shuffle_df['test_acc'][-10:-1]))
    weight_shuffle_acc.append(np.mean(weight_shuffle_df['test_acc'][-10:-1]))

    '''
    print(np.mean(acc_df['test_acc'][-10:-1]))
    print(np.mean(mask_shuffle_df['test_acc'][-10:-1]))
    print(np.mean(weight_shuffle_df['test_acc'][-10:-1]))
    '''
    
#     ax1.tick_params(axis='both', labelsize=tick_size)
#     ax2.tick_params(axis='both', labelsize=tick_size)

#     ax1.set_xlabel("Epoch ", fontsize=label_size)
#     ax1.set_ylabel("Test Accuracy", fontsize=label_size, color='C0')
#     ax2.set_ylabel("Model Sparsity", fontsize=label_size, color='C1')

#     #ax2.set_ylim([0, 55])
#     ax1.set_ylim([10, 95])
#     #ax1.set_xlim([0,152])
#     ax1.legend(fontsize=legend_size, loc='center left', bbox_to_anchor=(1.1, 0.8))#, loc='right')
#     ax1.grid(True)
#     #ax2.grid(True)
#     plt.savefig('hc_quantized_resnet20_cifar10.pdf', format='pdf', dpi=600, bbox_inches='tight', pad_inches=0.05)
```

```python
fig, ax1 = plt.subplots(figsize=(6, 4))
ax1.plot(imp_df['weight_remain'], imp_df['test_acc'], linewidth=line_width, c='C0', label='IMP')
ax1.plot(sp_list, vanilla_acc, linewidth=line_width, c='C0', label='HC, Suggested', linestyle='dashed')
ax1.plot(sp_list, weight_shuffle_acc, linewidth=line_width, c='C2', label='HC, weight shuffled', linestyle='dashed')
ax1.plot(sp_list, mask_shuffle_acc, linewidth=line_width, c='C1', label='HC, mask shuffled', linestyle='dashed')

plt.title("ResNet20 on CIFAR-10", fontsize=title_size-10)
#ax1.set_xlim([0,10])
ax1.invert_xaxis()
ax1.tick_params(axis='both', labelsize=tick_size)
ax1.set_xlabel("Fraction of unpruned weights", fontsize=label_size)
ax1.set_ylabel("Test Accuracy", fontsize=label_size)
ax1.legend(fontsize=legend_size, loc='center left', bbox_to_anchor=(1.1, 0.8))#, loc='right')
ax1.grid(True)
plt.savefig('hc_resnet20_cifar10.pdf', format='pdf', dpi=600, bbox_inches='tight', pad_inches=0.05)
```
